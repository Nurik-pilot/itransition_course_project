```python
%%pyspark
df_taxi = spark.read.table("analytics_lh_bronze.dbo.bronze_yellow_taxi")

from pyspark.sql.functions import col, year, month, dayofweek, hour, round

df_taxi_silver = df_taxi.dropDuplicates() \
    .filter(col("tpep_pickup_datetime").isNotNull()) \
    .filter(col("tpep_dropoff_datetime").isNotNull()) \
    .filter(col("trip_distance") > 0) \
    .filter(col("fare_amount") > 0) \
    .withColumn("pickup_year", year(col("tpep_pickup_datetime"))) \
    .withColumn("pickup_month", month(col("tpep_pickup_datetime"))) \
    .withColumn("pickup_day_of_week", dayofweek(col("tpep_pickup_datetime"))) \
    .withColumn("pickup_hour", hour(col("tpep_pickup_datetime"))) \
    .withColumn("trip_distance", round(col("trip_distance"), 2)) \
    .withColumn("fare_amount", round(col("fare_amount"), 2)) \
    .withColumn("total_amount", round(col("total_amount"), 2))

df_taxi_silver.write.mode("overwrite").saveAsTable("silver_yellow_taxi")
df_taxi_silver.count()
```

```python
%%pyspark
df_aq = spark.read.table("analytics_lh_bronze.dbo.bronze_air_quality")

from pyspark.sql.functions import col, to_timestamp, year, month, dayofweek, hour

df_aq_silver = df_aq.dropDuplicates() \
    .filter(col("value").isNotNull()) \
    .filter(col("value") >= 0) \
    .filter(col("datetime_utc").isNotNull()) \
    .withColumn("datetime_utc", to_timestamp(col("datetime_utc"))) \
    .withColumn("datetime_local", to_timestamp(col("datetime_local"))) \
    .withColumn("measure_year", year(col("datetime_utc"))) \
    .withColumn("measure_month", month(col("datetime_utc"))) \
    .withColumn("measure_day_of_week", dayofweek(col("datetime_utc"))) \
    .withColumn("measure_hour", hour(col("datetime_utc")))

df_aq_silver.write.mode("overwrite").saveAsTable("silver_air_quality")
df_aq_silver.count()
```

```python
%%pyspark
df_gdp = spark.read.table("analytics_lh_bronze.dbo.bronze_gdp")

from pyspark.sql.functions import col

df_gdp_silver = df_gdp.dropDuplicates() \
    .filter(col("gdp_usd").isNotNull()) \
    .filter(col("year").isNotNull()) \
    .withColumn("gdp_usd", col("gdp_usd").cast("double")) \
    .withColumn("year", col("year").cast("integer"))

df_gdp_silver.write.mode("overwrite").saveAsTable("silver_gdp")
df_gdp_silver.count()
```

```python
%%pyspark
df_fx = spark.read.table("analytics_lh_bronze.dbo.bronze_fx_rates")

from pyspark.sql.functions import col, to_date, year, month

df_fx_silver = df_fx.dropDuplicates() \
    .filter(col("usd_eur_rate").isNotNull()) \
    .filter(col("date").isNotNull()) \
    .withColumn("date", to_date(col("date"))) \
    .withColumn("usd_eur_rate", col("usd_eur_rate").cast("double")) \
    .withColumn("year", year(col("date"))) \
    .withColumn("month", month(col("date")))

df_fx_silver.write.mode("overwrite").saveAsTable("silver_fx_rates")
df_fx_silver.count()
```
